```python
# 각 품목에 대한 설명과 가격 정의
lovely_loveseat_description = "러블리 러브시트 - 나무 위에 폴리에스테르 테프트 블렌드로 제작된 매우 사랑스러운 소파. 높이 32인치 x 너비 40인치 x 깊이 30인치. 빨간색이든 흰색이든 사용 가능합니다."
lovely_loveseat_price = 254.00
```

```python
luxury_lamp_description = "럭셔리 램프 - 화려한 디자인과 최상의 조명을 제공하는 램프. 황금빛 프레임과 크리스털 몸체로 제작되었습니다. 높이 32인치."
luxury_lamp_price = 52.15
```

```python
# 판매세율 정의 (8.8%)
sales_tax = 0.088
```

```python
# 고객의 구매 내역 초기화
customer_1_itemization = ""
```

```python
# 첫 번째 고객의 총 지출 계산
customer_one_total = 0
customer_one_total += lovely_loveseat_price
customer_1_itemization += "러블리 러브시트: " + lovely_loveseat_description + "\n"

customer_one_total += luxury_lamp_price
customer_1_itemization += "럭셔리 램프: " + luxury_lamp_description + "\n"

```

```python
# 각 품목에 대한 판매세 계산
customer_lovely_loveseat_tax = lovely_loveseat_price * sales_tax
customer_luxury_lamp_tax = luxury_lamp_price * sales_tax
```

```python
# 고객의 총 판매세 계산
customer_total_tax = customer_lovely_loveseat_tax + customer_luxury_lamp_tax
```

```python
# 고객의 총 비용 계산
customer_total_cost = customer_one_total + customer_total_tax
```

```python
# 영수증 출력
print("고객 하나의 항목:")
print(customer_1_itemization)
print("Customer One Total:")
print(customer_total_cost)
```
